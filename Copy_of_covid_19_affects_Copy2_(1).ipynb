```python
age_mapping = {
    '19-25': 22,
    'Dec-18': 18,
    '33-40': 36.5,
    '60+': 65,
    '26-32': 29,
    '40-50': 45,
    '50-60': 55
}
```

```python
change_columns = ['prod_inc', 'sleep_bal', 'new_skill', 'fam_connect', 'relaxed', 'self_time']
df['lifestyle_change'] = df[change_columns].apply(lambda row: np.any(np.abs(row) > 0.5), axis=1).astype(int)
```

```python
df['lifestyle_change'].unique()
```

```python
df['lifestyle_change'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
numerical_transformer = StandardScaler()
```

```python
df_numerical = pd.DataFrame(numerical_transformer.fit_transform(df[numerical_columns]), columns=numerical_columns)
```

```python
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
```

```python
df_categorical = pd.get_dummies(df[object_columns], columns=object_columns)
```

```python
df_preprocessed = pd.concat([df_numerical, df_categorical, df['lifestyle_change']], axis=1)
```

```python
df_preprocessed.columns
```

```python
X = df_preprocessed.drop(columns='lifestyle_change')
y = df_preprocessed['lifestyle_change']
```

```python
X.columns
```

```python
X.to_csv('covid_preprocessed_data.csv', index=False)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify = y, random_state=0)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_dt = DecisionTreeClassifier()
```

```python
model_dt.fit(X_train, y_train)
```

```python
y_pred = model_dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_dt = accuracy_score(y_test, y_pred)
```

```python
print("Accuracy: ", accuracy_dt * 100)
print("\nClassification Report: \n", classification_report(y_test, y_pred))
```
